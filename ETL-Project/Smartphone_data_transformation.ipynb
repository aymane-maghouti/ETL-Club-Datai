### drop the NA value

### Convert the type of some features to numeric type

```python
df = df[df['screen_size'].apply(lambda x: x.isnumeric() or x.replace('.', '').isnumeric())]
```

```python
df['screen_size'] = pd.to_numeric(df['screen_size'])
df['ram']=pd.to_numeric(df['ram'])
df['rom']=pd.to_numeric(df['rom'])
df['battary']=pd.to_numeric(df['battary'])
df['previous_price']=pd.to_numeric(df['previous_price'])
df['current_price']=pd.to_numeric(df['current_price'])
```

```python
df.info()
```

```python
df.head()
```

```python
data = pd.DataFrame(df[['brand','screen_size','ram','rom','sim_type','battary','previous_price','current_price']])
```

```python
data.head()
```

```python
data.shape
```

### Adding a Mean Price Column

```python
data['price'] = (data['current_price'] + data['previous_price']) / 2
```

```python
data.head()
```

```python
data = pd.DataFrame(data[['brand','screen_size','ram','rom','sim_type','battary','price']])
```

```python
data.head()
```

### save the dataframe as csv file

```python
data.to_csv("cleaned_data/smartphone_Data.csv")
```
